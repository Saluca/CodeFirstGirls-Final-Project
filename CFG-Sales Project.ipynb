```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
sales = pd.read_csv('sales.csv')
print(sales.head())
```

```python
sale_list = sales['sales'].tolist()
print(sale_list)
```

```python
total_sales = sales['sales'].sum()
print('The total amount of money across all months is {}' .format(total_sales))
```

```python
change = sales['sales'].pct_change()
print(change)
```

```python
print(sales.describe())
```

```python
## Covariance matrix

sale_vs_expense = np.cov(sales.sales,sales.expenditure)
print(sale_vs_expense)
```

```python
# Pearson's correlation coefficient shows the strength and direction of correlation
from scipy.stats import pearsonr
pears_corr, p=pearsonr(sales.sales,sales.expenditure)
print(pears_corr)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))
fig.subplots_adjust(hspace=0.2, wspace=0.5)
sns.barplot(data=sales, x='month',y = 'sales',ax = axes[0] )
sns.barplot(data=sales,x='month',y='expenditure', ax = axes[1])
plt.show()
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,10))
ax1.pie(sales['sales'], labels = sales['month'], autopct= '%0.1f%%',pctdistance=0.85, labeldistance=1.1)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
ax2.pie(sales['expenditure'],labels = sales['month'],  autopct= '%0.1f%%', pctdistance=0.85, labeldistance=1.1)
ax1.set_title('Sales')
ax2.set_title('Expenditure')
plt.subplots_adjust(top=0.5, bottom=0.15)
plt.show()
```
